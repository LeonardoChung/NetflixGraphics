```python
# carregar bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# passar dataset a um dataframe
df = pd.read_csv("Netflix Engagement Dataset.csv")
display(df)
```

## a) Gráfico de linha

```python
# Gráfico de linha -> Tempo médio de visualização diário x Duração da assinatura

df_linha = df.groupby("Subscription Length (Months)")["Daily Watch Time (Hours)"].mean()

sns.set_theme(style="whitegrid")
plt.plot(df_linha.index, df_linha.values, color="#98FBCB")
plt.xlabel("Duração assinatura (em meses)")
plt.ylabel("Tempo médio visualização diário (em horas)")

plt.show()
```

O gráfico de linha apresentado em relação a Tempo Médio de Visualização Diário x Duração da Assinatura demonstra várias oscilações. Portanto, é possível que essas duas variáveis não possuam relação direta.

## b) Gráfico de área

```python
# Gráfico de área -> Engajamento Médio (1-10) por região
df_area = df.groupby("Region")["Engagement Rate (1-10)"].mean().sort_values()

sns.set_theme(style="whitegrid")
plt.fill_between(df_area.index, df_area.values, color="pink", alpha=0.6)
plt.xlabel("Região")
plt.ylabel("Engajamento médio (1-10)")

plt.show()
```

Pelo gráfico de área sobre Engajamento Médio por Região, observa-se que a América do Sul é a região com menor média de engajamento, enquanto a América do Norte é a com a maior. Repara-se também que as médias de todas as regiões ficam entre 5.3 e 5.6, com pouca variação entre elas.

##d) Histograma

```python
# Histograma -> Satisfação do Cliente x Taxa de Engajamento

df[["Engagement Rate (1-10)", "Customer Satisfaction Score (1-10)"]].plot.hist(alpha=0.5, bins=10, figsize=(10, 6), color=['#FF9671', '#4DABF5'])
plt.title('Satisfação do Cliente vs Taxa de Engajamento', fontsize=14)
plt.xlabel('Pontuação (1-10)', fontsize=12)
plt.ylabel('Frequência', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(['Satisfação do Cliente', 'Taxa de Engajamento'])

plt.tight_layout()
plt.show()
```

Os dados mostram que muitos usuários da Netflix continuam assistindo conteúdo mesmo quando não estão totalmente satisfeitos com o serviço.

## e) Gráfico de dispersão

```python
# Gráfico de dispersão -> Tempo de visualização diário x Idade

plt.scatter(df["Daily Watch Time (Hours)"], df["Age"], alpha=0.5, color="pink")
plt.xlabel("Tempo de visualização diário (em horas)")
plt.ylabel("Idade")

plt.show()
```

O gráfico de dispersão em relação a Tempo de Visualização Diário x Idade não demonstra algum trajeto ou padrão claro. Os pontos estão todos espalhados. Deste modo, é possível que essas duas variáveis não possuam relação direta.

##g) Gráfico de Pizza

```python
#Gráfico de pizza -> Dispositivos mais usados

dispositivo = df['Device Used Most Often'].value_counts()
dispositivo.plot.pie(autopct='%1.1f%%')

plt.title('Dispositivos mais usados')
plt.show()
```

O gráfico de pizza dos dispositivos mais usados demonstra que a divisão é bem equibilibrada.
